```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns

%matplotlib inline
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
import yfinance as yf
 
```

```python
#BANK OF AMERICA
BAC=yf.download('BAC',start,end)

#CITI BANK
C=yf.download('C',start,end)

#Goldman Sachs
GS=yf.download('GS',start,end)

#J.P MORGAN
JPM=yf.download('JPM',start,end)

#Morgan Stanley
MS=yf.download('MS',start,end)

#Wells Fargo
WFC=yf.download('WFC',start,end)
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
MS.head()
```

```python
#Creating a list of symbols/ticks.
tickers=['BAC','C','GS','JPM','MS','WFC']
```

```python
#Concatinating all data frame into one.
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

```python
bank_stocks.columns.names=('Bank Name','Stock Info')
```

```python
#Added column names
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns=pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for name in tickers:
    returns[name+' Returns']=bank_stocks[name]['Close'].pct_change()

#To drop the first value as it will be null.
returns.dropna()
```

** Creating a pairplot using seaborn of the returns dataframe for a overall idea.

```python
sns.pairplot(data=returns)
```

```python
#Worst day for each bank.
pd.concat([returns.idxmin(),returns.min()],axis=1)
```

```python
#Best day for each bank.
pd.concat([returns.idxmax(),returns.max()],axis=1)
```

```python
#To find the stock which deviates a lot.
returns.std()#Citi Bank is the riskiest
```

```python
sns.displot(returns['2009-01-01':'2009-12-31']['C Returns'])
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008.
